```python
def find_missing_percent(data):
    miss_df = pd.DataFrame({'ColumnName':[],'TotalMissingVals':[],'PercentMissing':[]})
    for col in data.columns:
        NumOfMissing = data[col].isnull().sum()
        if(NumOfMissing == 0):
            continue
        PercentOfMissing = (NumOfMissing / data.shape[0]) * 100
        data_dict = {"ColumnName" : col, "TotalMissingVals" : NumOfMissing, "PercentMissing" : PercentOfMissing}
        miss_df = miss_df.append(data_dict, ignore_index = True)
    return miss_df
```

```python
# Suggestion : If we remove all outlier, it is better to use mean not median
def median_imputation(data, missing_numeric_cols):
    for col in missing_numeric_cols:
        medianV = data[col].median()
        data[col] = data[col].fillna(medianV)
    return data

def mode_imputation(data, missing_categoric_cols):
    for col in missing_categoric_cols:
        modeV = data[col].mode().iloc[0]
        data[col] = data[col].fillna(modeV)
    return data
```

```python
def Model_Imputation(data, predictor_cols, target_cols):
    data_model_imputation = data.copy()
    data_numeric_predictors = data[predictor_cols].copy()
    
    for col in target_cols:
        missing_rows_index = data_model_imputation[col][data_model_imputation[col].isnull()].index
        X = data_numeric_predictors.copy()
        y = data_model_imputation[[col]]
        
        X_drop = X.drop(missing_rows_index, axis = 0)
        y_drop = y.drop(missing_rows_index, axis = 0)
        
        X_drop = X_drop.values
        y_drop = y_drop.values
        
        xgb = xgboost.XGBRegressor(objective="reg:squarederror", random_state=42)
        xgb.fit(X_drop, y_drop)
        y_predict = xgb.predict(X.values)
        
        data_model_imputation[col][missing_rows_index] = y_predict[missing_rows_index]
    return data_model_imputation
```

```python
corr = data.corr()
components = []
visited = set()
for col in corr.columns:
    if col in visited: continue
    component = set([col, ]) # each col is set if using set(col) it will be set of letters in col
    just_visited = [col, ]
    visited.add(col)
    while just_visited:
        c = just_visited.pop(0)
        for idx, val in corr[c].items():
            if abs(val) > 0.999 and idx not in visited:
                just_visited.append(idx)
                visited.add(idx)
                component.add(idx)
    components.append(component)

for component in components:
    plt.figure(figsize=(12, 8))
    sns.heatmap(corr.loc[component, component], cmap="Reds")
        
     
```

```python
def filter_corr(data, corr_value=0.8, vis=True):
    from copy import deepcopy
    """[summary]
    Filter pair of columns that have high correlation
    Args:
        data ([pd.DataFrame]): [Dataframe which need to compute and filter columns has high correlation]
        corr_value (float, optional): [take columns which has correlation value equal or above this value]. Defaults to 0.8.

    Returns:
        [tuple]: [tuple contain a series which has pair of columns along with their correlation value and dataframe (similar to correlation dataframe)]
    """
    def swap(_tup):
        tup = deepcopy(_tup)
        tup = list(tup)
        tup[0], tup[1] = tup[1], tup[0]
        return tup

    corr = data.corr()
    sort_corr = corr.unstack().sort_values(kind="quicksort", ascending=False)
    sort_corr = sort_corr[(abs(sort_corr.values) >= corr_value) & (abs(sort_corr.values) != 1)]
    col = [c[0] for c, _ in sort_corr.items()]
    idx = [c[1] for c, _ in sort_corr.items()]
    df_c = pd.DataFrame(data=np.zeros((len(sort_corr), len(sort_corr))), columns=col, index=idx)
    for k, _ in sort_corr.items():
        k1, k2 = k
        df_c.loc[k1, k2] = sort_corr[k]
        df_c.loc[k1, k1] = 1
    
    res = []
    for idx, (k, v) in enumerate(sort_corr.items()):
        k = list(k) + [v]
        if not res:
            res.append(k)
        if k not in res and swap(k) not in res:
            res.append(k)
    return res

pair = filter_corr(data)
pair
```
